```python
# import libraries
import os
import pandas as pd
import re
import string
import numpy as np

import nltk
nltk.download('vader_lexicon')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

import gensim
from gensim import corpora
from gensim.models import CoherenceModel
from gensim.models import LdaModel

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.decomposition import LatentDirichletAllocation

import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from scipy.cluster.hierarchy import dendrogram
```

# Data Extraction

```python
# Function to load data from a folder and creating a DataFrame
def load_lyrics_data(folder):
    songs = []
    for song_file in os.listdir(folder):
        if song_file.endswith(".txt"):
            file_path = os.path.join(folder, song_file)
            with open(file_path, 'r', encoding='utf-8') as file: #latin-1
                lyrics = file.read()
                songs.append({"Songs": os.path.splitext(song_file)[0], "Lyrics": lyrics})
    return pd.DataFrame(songs)

# Load training data
songs_folder = "lyrics"
df = load_lyrics_data(songs_folder)

# Load testing data
new_songs_folder = "new_lyrics"
new_df = load_lyrics_data(new_songs_folder)
```

# Data Cleaning

```python
print(f"Number of songs before cleaning: {df.shape[0]}")

# Dealing with missing data
# Find files with 'Lyrics' that appear blank after removing whitespace
blank_or_newline = df[df['Lyrics'].str.strip().eq('')]

print(f"Number of blank files: {len(blank_or_newline)}")

# Get the indices of rows to remove
indices_to_remove = blank_or_newline.index

# Remove the rows from the DataFrame
df.drop(indices_to_remove, inplace=True)

# Remove songs with the same lyrics
df.drop_duplicates(subset='Lyrics', inplace=True)

print(f"Number of songs after cleaning: {df.shape[0]}")
```

# Exploratory Data Analysis

## How many of the songs feature at least one pair of lines that rhyme?

```python
# Convert 'Songs' and 'Lyrics' columns to lowercase for case-insensitive matching
df['Songs'] = df['Songs'].str.lower()
df['Lyrics'] = df['Lyrics'].str.lower()

# Function to find rhyming lines in lyrics
def find_rhyming_lines(lyrics):
    # Split the lyrics into lines
    lines = lyrics.split('\n')
    
    # Iterate through the lines and check for rhyming pairs
    for i in range(len(lines) - 1):
        line1, line2 = lines[i], lines[i + 1]
        
        # Extract words from each line
        words1, words2 = re.findall(r'\w+', line1), re.findall(r'\w+', line2)
        
        # Check if there are words in both lines
        if words1 and words2:
            
            # Extract the last words
            word1, word2 = words1[-1].lower(), words2[-1].lower()
            
            # Check if words are ending with the same 2 letters
            if word1.endswith(word2[-2:]) or word2.endswith(word1[-2:]):
                return True
    
    return False

# Create a new column 'Rhyming Lines'
df['Rhyming Lines'] = df['Lyrics'].apply(find_rhyming_lines)

# Count the number of songs with at least one pair of rhyming lines
num_songs_with_rhyming_lines = df['Rhyming Lines'].sum()

print(f"Number of songs with at least one pair of rhyming lines: {num_songs_with_rhyming_lines}")
```

## How many of the songs feature the song name (found in the file name) in the song lyrics?

```python
# Preprocess the text
# Replace the dash in song names with space
df['Cleaned_Songs'] = df['Songs'].apply(lambda text: text.replace('-', ' '))

# Lowercase, remove whitespace, punctuation
df['Cleaned_lyrics'] = df['Lyrics'].apply(lambda text: text.lower())
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lambda text: text.replace('\n', ' '))
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lambda text: text.translate(str.maketrans('', '', string.punctuation)))
```

```python
# Create a new column 'Song Name in Lyrics' of boolean values based on whether the song name is found in the lyrics of that row.
df['Song Name in Lyrics'] = df.apply(lambda row: row['Cleaned_Songs'] in row['Cleaned_lyrics'], axis=1)

# Count the number of songs where the song name is in the lyrics
num_songs_with_name_in_lyrics = df['Song Name in Lyrics'].sum()

print(f"Number of songs that feature the song name in the lyrics: {num_songs_with_name_in_lyrics}")
```

## Sentiment Analysis

```python
# Preprocess the text
# Remove stopwords and tokenize the text
stop_words = stopwords.words('english')
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lambda words: word_tokenize(words))
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lambda words: [word for word in words if word not in stopwords.words('english')])
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lambda words: ' '.join(words))
```

```python
# Create a SentimentIntensityAnalyzer object
sia = SentimentIntensityAnalyzer()

# Function to calculate sentiment score for a given text
def get_sentiment_score(text):
    sentiment = sia.polarity_scores(text)
    compound_score = sentiment['compound']
    
    # Classify the compound score as positive, negative, or neutral
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each song's lyrics
df['Sentiment'] = df['Cleaned_lyrics'].apply(get_sentiment_score)

# Count the number of songs in each sentiment category and sort by counts
sentiment_counts = df['Sentiment'].value_counts().sort_index()
```

```python
# Display the results
# Bar charts of the number of songs in each sentiment category
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='pastel', order=['Negative', 'Neutral', 'Positive'])
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Number of Songs by Sentiment')
plt.xticks(rotation=45)

# Add numbers above the bars
for i, count in enumerate(sentiment_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()

# Function to generate and display a word cloud
def generate_word_cloud(text, title):
    stopwords_set = set(stopwords.words("english"))
    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords_set).generate(text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Generate and display word clouds for each sentiment category
negative_lyrics = " ".join(df[df['Sentiment'] == 'Negative']['Cleaned_lyrics'])
neutral_lyrics = " ".join(df[df['Sentiment'] == 'Neutral']['Cleaned_lyrics'])
positive_lyrics = " ".join(df[df['Sentiment'] == 'Positive']['Cleaned_lyrics'])

generate_word_cloud(negative_lyrics, 'Negative Sentiment')
generate_word_cloud(neutral_lyrics, 'Neutral Sentiment')
generate_word_cloud(positive_lyrics, 'Positive Sentiment')
```

# Classify the songs into topic clusters

## LDA

```python
# Preprocess the text
# Initialize NLTK's WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    text = ' '.join(tokens)
    return text

# Lemmatize the 'Lyrics' column
df['Cleaned_lyrics'] = df['Cleaned_lyrics'].apply(lemmatize_text)

# Create a tokenized lyrics column
df['Tokenized_Lyrics'] = df['Cleaned_lyrics'].apply(lambda words: word_tokenize(words))

# Create a dictionary representation of the documents
dictionary = corpora.Dictionary.load("dictionary")
# Use the two lines below instead for the first time
# dictionary = corpora.Dictionary(df['Tokenized_Lyrics'])
# dictionary.save("dictionary")

# Convert the tokenized documents into a Document-Term Matrix (DTM)
corpus = [dictionary.doc2bow(lyrics) for lyrics in df['Tokenized_Lyrics']]
```

### Find the number of topics

```python
# Perplexity and Coherence Scores:
# Lists to store scores
num_topics_list = []
perplexity_scores = []
coherence_scores = []

# Iterate through different numbers of topics
for num_topics in range(1, 11):
    lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)
    coherence_model_lda = CoherenceModel(model=lda_model, texts=df['Tokenized_Lyrics'], dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model_lda.get_coherence()
    perplexity_score = lda_model.log_perplexity(corpus)
    num_topics_list.append(num_topics)
    perplexity_scores.append(perplexity_score)
    coherence_scores.append(coherence_score)
    print(f"Nmber of topics: {num_topics}, Perplexity score: {perplexity_score}, Coherence score: {coherence_score}")

# Plot the scores
plt.figure(figsize=(10, 6))
plt.plot(num_topics_list, perplexity_scores, label='Perplexity', marker='o')
plt.plot(num_topics_list, coherence_scores, label='Coherence', marker='o')
plt.xlabel('Number of Topics')
plt.ylabel('Score')
plt.legend()
plt.show()
```

I would choose either 5 or 6 topics based on the balance between perplexity and coherence:

5 topics: It has a relatively low perplexity score and a good coherence score.

6 topics: It has a slightly higher perplexity score but the highest coherence score, indicating more interpretable topics.

### Build LDA model

```python
# Build the LDA model
num_topics = 5
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15, random_state=32)

# Create a new column 'Topics' in the DataFrame to store the assigned topics
df['Topics'] = [max(lda_model.get_document_topics(corpus[i]), key=lambda x: x[1])[0] for i in range(len(corpus))]

# Display the songs and their assigned topics
print(df[['Songs', 'Topics']])
```

### Visualize and interpret the topics

```python
# Extract topic probabilities for each document
topic_probs = np.zeros((len(corpus), num_topics))
for idx, doc_bow in enumerate(corpus):
    for topic, prob in lda_model.get_document_topics(doc_bow):
        topic_probs[idx, topic] = prob

# Dimensionality reduction using t-SNE
tsne_model = TSNE(n_components=2, perplexity=5, random_state=42)
tsne_data = tsne_model.fit_transform(topic_probs)

# Plotting the 2D t-SNE representation
plt.scatter(tsne_data[:, 0], tsne_data[:, 1], c=np.argmax(topic_probs, axis=1), cmap='viridis', marker='o', s=100)
plt.title('t-SNE Visualization of Topic Clusters')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```

```python
# Print the topics and their top words
topics = lda_model.print_topics(num_words=10)
for topic_num, topic in topics:
    print(f"Topic {topic_num}: {topic}")
```

**Topic 0: Love and Affection**
   - This topic seems to revolve around themes of love and affection. Words like "love," "im," "know," "want," "got," and "girl" suggest a strong emphasis on love and romantic relationships. It conveys feelings of desire and affection.

**Topic 1: Actions and Expressions**
   - This topic appears to be about actions and expressions. Words like "na," "let," "gon," "see," and "together" suggest a focus on taking action and expressing oneself. It might involve planning or making decisions.

**Topic 2: Assertiveness and Statements**
   - This topic seems to involve assertiveness and statements. Words like "dont," "back," "na," "get," "say," and "wont" indicate a tone of making statements or expressing one's perspective with confidence.

**Topic 3: Positive Emotions and Affirmations**
   - This topic is related to positive emotions and affirmations. Words like "know," "yeah," "baby," "oh," and "want" indicate a positive and enthusiastic tone. It might involve expressions of joy, affirmation, or optimism.

**Topic 4: Emotions and Expression**
   - This topic could be about emotions and expressions. Words like "going," "come," "cry," "lonely," and "mine" suggest a focus on emotions and expressions, possibly related to personal experiences and feelings.

```python
# Function to visualize topics using word clouds
def visualize_topics(lda_model, num_topics):
    fig, axes = plt.subplots(1, num_topics, figsize=(16, 8))
    for i, ax in enumerate(axes):
        topic_words = lda_model.show_topic(i, topn=10)
        wordcloud = WordCloud(width=400, height=200, background_color='white').generate_from_frequencies(dict(topic_words))
        ax.imshow(wordcloud, interpolation='bilinear')
        ax.set_title(f"Topic {i}")
        ax.axis('off')
    plt.show()

visualize_topics(lda_model, num_topics=num_topics)
```

```python
# Initialize a list to store predicted topics
predicted_topics = []

# Iterate over the rows of new_df and predict topics
for index, row in new_df.iterrows():
    doc = new_corpus[index]
    predicted_topic = max(lda_model.get_document_topics(doc), key=lambda x: x[1])[0]
    predicted_topics.append(predicted_topic)

# Assign the list of predicted topics to the DataFrame
new_df["Predicted Topic"] = predicted_topics

# Display the results
print(new_df[["Songs", "Predicted Topic"]])
```
